# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([1,2])
ypoints = np.array([5,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,20])
ypoints = np.array([10,200])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,8])
ypoints = np.array([9,23])

plt.plot(xpoints, ypoints, 'x')
plt.show()
```

```python
ypoints = np.array([3,4,2,4])

plt.plot(ypoints, '*''g')
plt.show()
```

```python
xpoints = np.array([1,2,4,10])
ypoints = np.array([9,23, 23, 4])

plt.plot(xpoints, ypoints, marker='*', ms=10, color='red')
plt.show()
```

```python
ypoints = np.array([3,5,2,34])
plt.plot(ypoints, marker = 'o', ms = 10, mec = 'r', color='black')
plt.show()
```

```python
ypoints = np.array([3,5,2,4,3])
plt.plot(ypoints, linestyle = 'dashed', marker = 'o', ms = 20, mec = 'hotpink', mfc = 'green', linewidth= 10)
plt.show()
```

```python
y1 = np.array([3, 4, 2, 12])
y2 = np.array([4, 2, 32, 17])

plt.plot(y1, color='red')
plt.plot(y2, 'k')
plt.show()
```

```python
x = np.array([35, 42, 23, 12])
y = np.array([42, 23, 32, 17])

plt.plot(x, y, color='r', linewidth = 1)

plt.xlabel("Input", fontsize=20, color='r')
plt.ylabel("Output", fontsize=20, color='g')
plt.title("sample", fontsize = 30)
plt.grid(color="green", linestyle="dotted")
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5, 10, 15, 20, 25])
plt.scatter(x,y)
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([42, 23, 32, 17])

plt.bar(x, y, color = "black")
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([42, 23, 32, 17])

plt.barh(x, y, color = "green")
plt.show()
```

```python
x = np.random.normal(12,5,500)
plt.hist(x)
plt.show()
```

```python
y = np.array([10, 343, 34, 54, 34])
plt.pie(y)
plt.show()
```

```python
y = np.array([34, 23, 98, 8])
myLabels = ["Cherries", "Apples", "Bananas", "Guavas"]

plt.pie(y, labels = myLabels, colors=["red", "orange", "yellow", "green"])
plt.show()
```

```python
y = np.array([34, 23, 98, 8])
myLabels = ["Cherries", "Apples", "Bananas", "Guavas"]

plt.pie(y, labels = myLabels, startangle = 180, colors=["red", "orange", "yellow", "green"])
plt.show()
```

```python
y = np.array([34, 23, 98, 8])
myLabels = ["Cherries", "Apples", "Bananas", "Guavas"]
myexplode = [0, 0.95, 0.8, 0.2]

plt.pie(y, labels = myLabels, explode = myexplode, colors=["red", "orange", "yellow", "green"])
plt.show()
```

```python
y = np.array([34, 23, 98, 8])
myLabels = ["Cherries", "Apples", "Bananas", "Guavas"]

plt.pie(y, colors=["red", "orange", "yellow", "green"])
plt.legend(labels=myLabels)
plt.show()
```
