# Seaborn

```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Yhills_July12_Analytics/main/titanic-training-data.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
sns.relplot(x='Pclass', y='Fare', hue = 'Sex', data = df)
```

```python
sns.catplot(x='Embarked', y='Age', hue = 'Sex', data = df)
```

```python
sns.catplot(x='Pclass', y='Age', hue = 'Sex', data = df)
```

```python
sns.catplot(x='Pclass', y='Age', jitter = False, hue = 'Sex', data = df)
```

```python
sns.catplot(x='Pclass', y='Age', kind = 'swarm', hue = 'Sex', data = df, s=10)
```

```python
sns.catplot(x='Pclass', y='Age', kind = 'box', hue = 'Sex', data = df)
```

```python
sns.catplot(x='Pclass', y='Age', kind = 'violin', hue = 'Sex', data = df)
```

```python
sns.catplot(x='Pclass', y='Age', kind = 'violin', split = True, hue = 'Sex', data = df)
```

```python
sns.catplot(x='Pclass', y='Age', kind = 'point', hue = 'Sex', data = df)
```
